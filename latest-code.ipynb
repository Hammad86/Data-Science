```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.remote.webelement import WebElement
from selenium.webdriver.common.action_chains import ActionChains
import csv

import time
```

```python
with open("quizLinks.csv") as f:
    data = csv.reader(f)
    for i in data:
        for j in i:
            try:
                driver = webdriver.Chrome(executable_path='chromedriver.exe')
                driver.maximize_window()
                driver.get(j)
#                 time.sleep(5)
                element = WebDriverWait(driver, 40).until(
                        EC.element_to_be_clickable((By.XPATH, "//button[@class='UISubmit-component is-loaded QuizCover-start-button']")) )
                element.click()
            except:
                continue

            for x in range(1,11):
                try:
                    allOptions = WebDriverWait(driver, 40).until(
                                        EC.presence_of_all_elements_located((By.XPATH, "//span[@class='QuizQuestionOption-item-label']")) )
                    option = WebDriverWait(driver, 40).until(
                                        EC.element_to_be_clickable((By.XPATH, "//div[@class='QuizQuestionOption-item-icon']")) )

                    # print(option.text)
                    option.click()
                    nextBtn = WebDriverWait(driver, 40).until(
                                        EC.element_to_be_clickable((By.XPATH, "//button[@class='QuizQuestion-feedback-header-action']")) )

                    correctOption = ""
                    incorrectOption = []
                except:
                    continue
                    
                for y in allOptions:
            #         print('is-correct' in i.get_attribute('class'))
                    if 'is-correct' in y.get_attribute('class'):
                        correctOption = y.text

                    else:
                        incorrectOption.append(y.text)

                question = WebDriverWait(driver, 50).until(
                                    EC.presence_of_element_located((By.XPATH, "//h2[@class='QuizQuestion-info-title']")) )
            #     time.sleep(10)
                with open('quizData.csv', 'a+', encoding='utf-8-sig', newline='') as f:
                        writer = csv.writer(f)
                        if len(incorrectOption) == 3:
                            writer.writerow([question.text,correctOption,incorrectOption[0],incorrectOption[1],incorrectOption[2]])
                        elif len(incorrectOption) == 2:
                            writer.writerow([question.text,correctOption,incorrectOption[0],incorrectOption[1]])
                        elif len(incorrectOption) == 1:
                            writer.writerow([question.text,correctOption,incorrectOption[0]])

                print(question.text) 
                print(correctOption)        
                print(incorrectOption)
#                 time.sleep(5)
                nextBtn.click()
            driver.close()    
```
