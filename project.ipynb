```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.remote.webelement import WebElement
from selenium.webdriver.common.action_chains import ActionChains

import time
```

```python
# Open Webpage on ChromeDriver
driver = webdriver.Chrome(executable_path='chromedriver.exe')
driver.get('https://www.wizardingworld.com/quiz')

element = driver.find_element_by_xpath("//div[@class='HubMoreToExplore_more_to_explore__2EvM0']")

driver.execute_script("arguments[0].scrollIntoView();", element)


flag = True
Y = 200
X = 0
while flag:
    # Scroll till the fixed height
    driver.execute_script(f"window.scrollTo({X}, {Y})")
    time.sleep(2)
    X = Y
    Y = Y + 200
    quizLinks = driver.find_elements_by_xpath("//a[@class='ImageWithDetails_image_wrapper__25tsD']")
    for i in quizLinks:
        print(i.get_attribute('href'))
   
```
