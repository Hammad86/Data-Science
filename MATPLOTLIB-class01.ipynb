```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Creating line graph
x = np.linspace(0,10,1000) #linespace(starting,ending,total points between them)
plt.plot(x,np.sin(x)) #plot(x-axis , y-axis)
plt.plot(x,np.cos(x))
plt.show()
```

```python
#Creating subplot basic
plt.Figure()
plt.subplot(2,1,1) #subplot(rows,column,index)
plt.plot(x,np.sin(x))
plt.subplot(2,1,2)
plt.plot(x,np.cos(x))
plt.show()
```

```python
#Creating subplots(OOP)
fig,ax = plt.subplots(2)
ax[0].plot(x,np.sin(x))
ax[1].plot(x,np.cos(x))
plt.show()
```

```python
#Using Different graph style with colors and line styles
plt.style.use('seaborn-whitegrid')
x = np.linspace(0,10,1000)
plt.plot(x,np.cos(x-3),color='#000',linestyle='dashed',label='cos(x-3)')
plt.plot(x,np.cos(x),color='b',linestyle='dotted',label='cos(x)')
plt.plot(x,np.cos(x-1),color='g',linestyle=':' ,label='cos(x-1)')

#giving linestyle(-.) and color(yellow) in 1 step
plt.plot(x,np.cos(x-2),'-.y' ,label='cos(x-2)')

#labeling
plt.title('Cos Graph')
plt.xlabel('X')
plt.ylabel('cos(x)')

#legend shows the label of each line
plt.legend()
plt.show()
```
